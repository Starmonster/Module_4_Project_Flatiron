# Weighted Recommender as a solution to the cold start

We're going to use IMBDs weighted average formula to see how this affects our data in terms of how we describe a poplular movie.

### IMBD weighted average rating formula

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
from plotly.offline import init_notebook_mode, iplot
import plotly.graph_objs as go
init_notebook_mode(connected=True)
```

```python
df = pd.read_csv('./ml-latest-small/ratings.csv')
df.info()
```

```python
df.head()
```

```python
df.movieId.nunique()
```

```python
df.movieId.value_counts()
```

```python
df_ave = df.groupby('movieId').agg({'rating':[np.size, np.mean]})
df_ave = df_ave.reset_index()
```

```python
df_ave.head()
```

```python
df_ave.columns
```

```python
#Now we'll calculate all the components based on the formula given at the top of notebook

v = df_ave['rating', 'size'] #No of ratings per movie in dataset
R = df_ave['rating', 'mean'] #Average of the ratings, v
C = df_ave['rating', 'mean'].mean() #The mean average rating across the whol set
m = df_ave['rating', 'size'].quantile(0.7) #movies that are above the 70%ile for vote counts
```

```python
#Create a new column and fill values based on our formula output.
df_ave['weighted_average'] = ((R*v) + (C*m))/(v+m)
```

```python
df_ave.head()
```

```python
#Let's see the top weighted averages in descening order
sorted_ranks = df_ave.sort_values('weighted_average', ascending=False)
sorted_ranks
```

As we can see from the above data frame split, movies with a higher relative rating size have been awarded a good weighting near to the mean weighting. Movies with only a single rating have a lower weighting meaning there weighted average penalises their relatively small number of ratings. 

```python

fig, ax1 = plt.subplots(figsize=(12,6))

#plt.figure(figsize=(12,6))
ax1 = sns.barplot(x=sorted_ranks_1[('weighted_average', '')].head(5), y=sorted_ranks_1['title'].head(5), data=sorted_ranks_1)
plt.xlim(2, 5)
plt.title('Graph to show best and worst by weighted average ratings', fontsize=15)
plt.xlabel('Weighted Average Score', fontsize=12)
plt.ylabel('Movie Title', fontsize=12)
```
