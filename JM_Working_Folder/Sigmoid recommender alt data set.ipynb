```python
#rename the credits id column to match movies id column for ease of merge
cred_col_renamed = credits.rename(index=str, columns={'movie_id':'id'})

#merge the dataframes
df_merged = movies.merge(cred_col_renamed, on='id')

df_merged.head()
```

```python
#Now we drop some unused columns
movies_cleaned_df = df_merged.drop(columns=['homepage', 'title_x', 'title_y', 'status', 'production_countries'])
movies_cleaned_df.head()
```

## Make a recomendation

```python
movies_cleaned_df.head(1)['overview'].all()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
#We use Term Frequency, Innverse Document Frequency to determine the weights of each term in our dataframe
tfv = TfidfVectorizer(min_df=3, max_features=None, strip_accents='unicode', analyzer='word', token_pattern=r'\w{1,}', 
                    ngram_range=(1,3), stop_words='english')
```

```python
#We'll replace any NaNs with empty strings
movies_cleaned_df['overview'] = movies_cleaned_df['overview'].fillna('')
```

```python
#Fit the TF-IDF on the 'overview' text, and create our sparse matrix
tfv_matrix = tfv.fit_transform(movies_cleaned_df['overview'])
```

```python
tfv_matrix
```

```python
tfv_matrix.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

#create and instance of the sigmoid kernel class
sig = sigmoid_kernel(tfv_matrix, tfv_matrix) #compare how summaries are related to each other
```

```python
indices = pd.Series(movies_cleaned_df.index, index=movies_cleaned_df['original_title']).drop_duplicates()
```

```python
#Write a recommendation function

def give_rec(title, sig=sig):
    #get the index corresponding to the original title
    idx = indices[title]
    
    #Get the sigmoid pairwise similarity scores
    sig_scores = list(enumerate(sig[idx]))
    
    
    #sort the movies into with top recommendations at the top
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)
    
    #Narrow into the top 5 most similar
    sig_scores = sig_scores[1:6] #remember that index 0 is our input title!! 
    
    #Identify the movie indices corresponding to our sorted / filtered sig scores list
    movie_indices = [i[0] for i in sig_scores]
    
    #And finally we convert the indices into actual titles and return them
    return movies_cleaned_df['original_title'].iloc[movie_indices]
    
    
```

```python
give_rec('Jurassic Park')
```
